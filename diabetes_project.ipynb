# how to handle imbalance data

```python
# we can handle imbalance dataset by using sampling techniques
#1.imbalance data will observe on classification prob # 2.split our data for training using sampling techniques
#Sampling technique: it is used to split our data for training machine learning model
# types of sampling technique:1. under sampling 2.over sampling 3.smote technique
```

```python
#1.undersampling: it is technique used to balance imbalanced dataset by decreasing the size of majority daataset to minority dataset
# is it good to delete original data,nobcoztraining with large data gives more better result than less data
# so we will not prefer undersampling to handle imbalance data

#2.oversampling:it is used to balance imbalanced dataset by increasing the size of minority dataset to majority dataset
#is it good to add duplicate data? no bcoz same info we are giving  multiple times ,ML model will not trained
#therefore this technique also we wont prefer

#3.smote technique: Smote means synthetic minority over sampling technique(it does not contain duplicates records
#it is oversampling technique where synthetic samples(approximated data) are generated for minority data
```

```python
total_num_preg=len(df.loc[df['num_preg']==0])
print(f'number of rows missing in num_preg column is {total_num_preg}')

total_glucose_conc=len(df.loc[df['glucose_conc']==0])
print(f'number of rows missing in glucose_conc column is {total_glucose_conc}')

total_diastolic_bp=len(df.loc[df['diastolic_bp']==0])
print(f'number of rows missing in diastolic_bp column is {total_diastolic_bp}')

total_insulin=len(df.loc[df['insulin']==0])
print(f'number of rows missing in insulin column is {total_insulin}')

total_bmi=len(df.loc[df['bmi']==0])
print(f'number of rows missing in bmi column is {total_bmi}')

total_diab_pred=len(df.loc[df['diab_pred']==0])
print(f'number of rows missing in diab_pred column is {total_diab_pred}')

total_age=len(df.loc[df['age']==0])
print(f'number of rows missing in age column is {total_age}')
```

```python
# how to handle zeros in our data? we handle zeros by sckit-learn software,from impute module and import simple imputer class
# by using simple imputer we will replace zero(0) values with mean or mode
# it will handle both x axis and y axis......x-axis=0 and y-axis=1
```

```python
from sklearn.impute import SimpleImputer
fill_zeros=SimpleImputer(missing_values=0,strategy='mean')      # replace 0 with mean values
x_train=fill_zeros.fit_transform(x_train)             # replace 0 with meanvalue in x_train
x_test=fill_zeros.fit_transform(x_test)                # replace 0 with meanvalue in x_test
```

```python
# check rows weather 0 is replaced with mean
x_train[0:3]
```

# standardized the data

```python
# from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# step 5 : Training the model
## train the model using naive bayes algorithm

```python
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()  # creating object for Gaussian cls 
nb_model.fit(x_train,y_train.ravel()) # training a model 
# ravel() func used to convert 2D array into 1D array
# fit func is used to train a model by passing training data 
# y ravel() func used only in y_train? x_train contains multiple column but y_train contain only one column
```

```python
# now check proportion how much model is trained by using x_train

from sklearn import metrics  # metrics module which contains measurement related func            
nb_predict_model=nb_model.predict(x_train)
print(f'predicted accuracy {metrics.accuracy_score(y_train,nb_predict_model)}')  #y_train,nb_predict_train will compare original to predicted output

# O/P: TRAINING ACCURACY=75%
```

```python
#testing model by passing x_test data
nb_predict_test=nb_model.predict(x_test)
print(f'predicted accuracy {metrics.accuracy_score(y_test,nb_predict_test)}')

#O/P: TESTING ACCURACY: 73%
```

# formula
## accuracy score= ((TP+TN)/(TP+FP+FN+TN))*100

## how to know whether accuracy is crt or not? by using 1. confusion matrix 2.classification report

```python
print("confusion matrix")
print(f"{metrics.confusion_matrix(y_test,nb_predict_test)}")    #o/p:- [[tp fn]
                                                                 #      [fp tn]]
```

```python
print("classification report")
print(f"{metrics.classification_report(y_test,nb_predict_test)}")  
```

```python
# precesion: quality of prediction made by model i.e +ve prediction or -ve prediction
#Recall:it will give proportion of true  +ve or true -ve to total prediction
# F1 score: is the harmonic mean of precision and recall
```

# 2. Training ur model Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(random_state=42)  # create object for random forest
rf_model.fit(x_train,y_train.ravel())   # training a machine
```

```python
rf_predict_model=rf_model.predict(x_train)
print(f'predicted accuracy {metrics.accuracy_score(y_train,rf_predict_model)}')          # o/p :training accuracy is 100%
```

```python
#testing model by passing x_test data
rf_predict_test=rf_model.predict(x_test)
print(f'predicted accuracy {metrics.accuracy_score(y_test,rf_predict_test)}')            # o/p: testing accuracy is 75%
```

```python
print("confusion matrix")
print(f"{metrics.confusion_matrix(y_test,rf_predict_test)}")    #o/p:- [[tp fn]
                                                                 #      [fp tn]]
```

```python
print("classification report")
print(f"{metrics.classification_report(y_test,rf_predict_test)}")  
```
